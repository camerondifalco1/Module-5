### Analysis

1 Variability and Standard Deviation:
 
The drug "Ketapril" shows the highest tumor volume variance (68.55) and standard deviation (8.28), indicating that the tumor volumes observed during the trials with this drug have a wide spread around the mean. This could point to a less consistent response to the treatment among different subjects.

Conversely, "Capomulin" and "Ramicane" have the lowest variance and standard deviation, suggesting a more consistent response across the subjects treated with these drugs.


2 Tumor Volume Metrics Comparison:

"Capomulin" and "Ramicane" show the lowest mean and median tumor volumes, around 40. These numbers might suggest greater effectiveness in reducing tumor size compared to the other drugs in the study.

On the other hand, "Ketapril" has the highest mean tumor volume (55.24), possibly indicating a lesser effectiveness in reducing tumor size or even a potential tumor growth-promoting effect.


3 Standard Error of the Mean (Std. Err.):

The standard error of the mean is an indication of how accurately the sample mean represents the population mean. A lower standard error (such as in "Capomulin" and "Ramicane") suggests a more reliable estimation of the population mean from the sample data.

Drugs like "Ketapril" and "Naftisol" have higher standard errors, indicating that the means for these drugs may be less representative of the true population means. This could be due to various factors such as smaller sample size or higher variability within the data.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")

# Display the data table for preview
merged_df
```

```python
# Checking the number of mice.
total_mice = mouse_metadata['Mouse ID'].count()
total_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merged_df[merged_df.duplicated(subset= ['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
print('Duplicate Mice:', duplicate_mice)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_data = merged_df[merged_df['Mouse ID'].isin(duplicate_mice)]
print('Optional Data for the Duplicate Mice:')
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = mouse_metadata[~mouse_metadata['Mouse ID'].isin(duplicate_mice)]
print('Cleaned Data:')
cleaned_df
```

```python
# Checking the number of mice in the clean DataFrame.
total_mice = cleaned_df['Mouse ID'].count()
total_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
grouped_df = merged_df.groupby('Drug Regimen')['Tumor Volume (mm3)']

summary_df = pd.DataFrame({
    'Mean Tumor Volume': grouped_df.mean(),
    'Median Tumor Volume': grouped_df.median(),
    'Tumor Volume Variance': grouped_df.var(),
    'Tumor Volume Std. Dev.': grouped_df.std(),
    'Tumor Volume Std. Err.': grouped_df.sem()
})

summary_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_df = merged_df.groupby('Drug Regimen').agg({
    'Tumor Volume (mm3)':
    ['mean',
     'median',
     'var',
     'std',
     'sem']
})

summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
filtered_df = merged_df[merged_df['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Start by getting the last (greatest) timepoint for each mouse
grouped_df = filtered_df.groupby('Mouse ID')['Timepoint'].max()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume_df = pd.merge(grouped_df, merged_df, on=['Mouse ID', 'Timepoint'], how='left')
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    subset = final_tumor_volume_df[final_tumor_volume_df['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    
    # Add subset
    tumor_vol_data.append(subset)
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers
    quartiles = subset.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = subset[(subset < lower_bound) | (subset > upper_bound)]
    
    # Print results
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumor_vol_data, labels=treatments, flierprops = dict(marker = "s", markerfacecolor = "orange"))
ax1.boxplot(tumor_vol_data, labels=treatments, medianprops = dict(color = "blue", linewidth = 1.5))
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_df = merged_df[merged_df['Drug Regimen'] == 'Capomulin']

# Select a single mouse
mouse_id = capomulin_df['Mouse ID'].iloc[0]

# Filter the data to only include the rows that contain the selected mouse
mouse_df = capomulin_df[capomulin_df['Mouse ID'] == mouse_id]

# Create the line plot
plt.plot(mouse_df['Timepoint'],
        mouse_df['Tumor Volume (mm3)'],
        color="blue")
plt.title(f'Capomulin Treatment of Mouse {mouse_id}')
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_df = merged_df[merged_df['Drug Regimen'] == 'Capomulin']

# Group the data by Mouse ID and find the average tumor volume for Mouse ID
grouped_df = capomulin_df.groupby('Mouse ID').agg({'Tumor Volume (mm3)':
                                                  'mean',
                                                  'Weight (g)': 'first'})

# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
plt.scatter(grouped_df['Weight (g)'],
           grouped_df['Tumor Volume (mm3)'],
           color="blue")
plt.title('Mouse Weight vs. Average Observed Tumor Volume, Capomulin Regimen')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
from scipy.stats import linregress

# Filter the data to only include Capomulin rows
capomulin_df = merged_df[merged_df['Drug Regimen'] == 'Capomulin']

# Group the data by Mouse ID and calculate the average tumor volume for each Mouse ID
grouped_df = capomulin_df.groupby('Mouse ID').agg({'Tumor Volume (mm3)':
                                               'mean',
                                               'Weight (g)': 'first'})

# Calculate the correlation coefficient between mouse weight and average tumor volume
correlation = grouped_df['Weight (g)'].corr(grouped_df['Tumor Volume (mm3)'])
print(f"The Correlation between mouse weight and average tumor volume is {correlation}")

# Calculate the linear regression model
(slope, intercept, rvalue, pvalue, stderr) = linregress(grouped_df['Weight (g)'],
                                                       grouped_df['Tumor Volume (mm3)'])
regress_values = grouped_df['Weight (g)'] * slope + intercept
print(f"The linear regression equation is y = {slope}x + {intercept}")

# Create a scatter plot of mouse weight vs. average tumor volume for the Capomulin regimen
plt.scatter(grouped_df['Weight (g)'],
           grouped_df['Tumor Volume (mm3)'],
           color="blue")
plt.plot(grouped_df['Weight (g)'],
        regress_values, color="orange")
plt.title('Mouse Weight vs. Average Tumor Volume, Capomulin Regimen')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
